```python
import pandas as pd 

#Load the dataset
df=pd.read_csv("all_stocks_5yr.csv")

df.head()
```

```python
# 1.Checking basic info before cleaning
print("Before Cleaning:")
print(df.info())
print(df.isnull().sum())

# 2.Removing rows with missing values (NaN)
df = df.dropna()

# 3.Remove duplicate rows 
df = df.drop_duplicates()

# 4.Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# 5.Sort values by company name and date 
df = df.sort_values(by=['Name', 'date']).reset_index(drop=True)

# 6.Final check after cleaning
print("\nAfter Cleaning:")
print(df.info())
print(df.isnull().sum())
print(df.head())
```

```python
df.to_csv("cleaned_stocks.csv", index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('stock.csv')
print(df)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
print(df.head())
```

```python
print(df.columns)
```

```python
print(df.shape)
```

```python
df = df.sort_values(['Name','date']).reset_index(drop=True)
```

```python
df[['open','high','low']] = df.groupby('Name')[['open','high','low']].transform(lambda g: g.ffill().bfill())
```

```python
print("Missing values after fill:\n", df.isnull().sum())
```

```python
top_name = df['Name'].value_counts().idxmax()   # most frequent ticker
print("Modelling for:", top_name)

df_company = df[df['Name'] == top_name].copy().reset_index(drop=True)
print("Company rows:", df_company.shape[0])
df_company.head()
```

```python
print(df_company.info())
```

```python
print(df_company.describe())
```

```python
#Missing values heatmap
plt.figure(figsize=(6,2))
sns.heatmap(df_company.isnull(), cbar=False)
plt.title("Missing values")
plt.show()
```

```python
# Time-series plot of close price
plt.figure(figsize=(12,5))
plt.plot(df_company['date'], df_company['close'])
plt.title(f"{top_name} Close Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

```python
# Volume vs Close scatter
RANDOM_STATE = 42 # Define RANDOM_STATE
sample = df_company.sample(min(2000, len(df_company)), random_state=RANDOM_STATE)
plt.figure(figsize=(8,5))
plt.scatter(sample['close'], sample['volume'], alpha=0.3)
plt.xlabel("Close")
plt.ylabel("Volume")
plt.title("Close vs Volume (sample)")
plt.show()
```

```python
# Distribution of returns (will compute returns next, but show close dist)
plt.figure(figsize=(8,4))
sns.histplot(df_company['close'], bins=50, kde=True)
plt.title("Distribution of Close Price")
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Use close price series
series = df_company['close']

# Fit ARIMA(p,d,q) = (5,1,0) as a simple start
model = ARIMA(series, order=(5,1,0))
model_fit = model.fit()

# Print summary
print(model_fit.summary())

# Forecast next 30 days
forecast = model_fit.forecast(steps=30)
print(forecast)

# Plot actual vs forecast
plt.figure(figsize=(10,5))
plt.plot(series, label="Actual")
plt.plot(range(len(series), len(series)+30), forecast, label="Forecast")
plt.legend()
plt.title("ARIMA Forecast")
plt.show()
```

```python
# Basic forecast evaluation 
import numpy as np

# Use the last 30 points as "test"
train = series[:-30]
test = series[-30:]

# Re-fit ARIMA on training
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(train, order=(5,1,0))
model_fit = model.fit()

# Forecast 30 steps
forecast_result = model_fit.get_forecast(steps=30)
forecast_mean = forecast_result.predicted_mean
forecast_ci = forecast_result.conf_int()

# Calculate RMSE using numpy
rmse = np.sqrt(np.mean((test - forecast_mean)**2))
print("RMSE on last 30 days:", rmse)

# Plot actual vs forecast with confidence interval
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(series, label="Actual")
plt.plot(range(len(train), len(series)), forecast_mean, label="Forecast")
plt.fill_between(range(len(train), len(series)),
                 forecast_ci.iloc[:,0],
                 forecast_ci.iloc[:,1], color='pink', alpha=0.3)
plt.legend()
plt.title("ARIMA Forecast with Confidence Interval")
plt.show()
```

```python
# 1.Import SARIMAX 
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# 2.Fit a simple SARIMA model 
# (p,d,q) = (1,1,1), seasonal (P,D,Q,s) = (1,1,1,5)
# s=5 is just a simple example for seasonality
sarima_model = SARIMAX(series, order=(1,1,1), seasonal_order=(1,1,1,5))
sarima_fit = sarima_model.fit(disp=False)

# 3.Print summary 
print(sarima_fit.summary())

# 4.Forecast next 30 days 
sarima_forecast = sarima_fit.get_forecast(steps=30)
forecast_mean = sarima_forecast.predicted_mean
forecast_ci = sarima_forecast.conf_int()

# 5.Plot actual vs SARIMA forecast
plt.figure(figsize=(10,5))
plt.plot(series, label="Actual")
plt.plot(range(len(series), len(series)+30), forecast_mean, label="SARIMA Forecast", color='orange')
plt.fill_between(range(len(series), len(series)+30),
                 forecast_ci.iloc[:,0],
                 forecast_ci.iloc[:,1], color='lightgreen', alpha=0.3)
plt.legend()
plt.title("SARIMA Forecast (Simple Example)")
plt.show()
```

```python
#  Prophet
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Prepare data
prophet_data = df_company['close'].reset_index()
prophet_data.columns = ['ds', 'y']  # Prophet needs 'ds' (date) and 'y' (value)

# Convert to datetime, drop invalid rows
prophet_data['ds'] = pd.to_datetime(prophet_data['ds'], errors='coerce')
prophet_data = prophet_data.dropna()

# Fit simple Prophet model
model = Prophet(daily_seasonality=True)  # simple beginner-friendly
model.fit(prophet_data)

# Forecast next 30 days
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

# Plot forecast
model.plot(forecast)
plt.title("Prophet Forecast (Simple, Beginner-Friendly)")
plt.show()
```
